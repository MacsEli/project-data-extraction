```python
##Selenium
driver = webdriver.Chrome("C:/Users/MacsEli/Downloads/Chrome/chromedriver.exe")
driver.get("https://es.tradingview.com/ideas/search/APPLE/")
time.sleep(10)
ActionChains(driver).send_keys(Keys.END).perform()
time.sleep(5)
ActionChains(driver).send_keys(Keys.END).perform()
time.sleep(5)
ActionChains(driver).send_keys(Keys.END).perform()
time.sleep(5)
page_source = driver.page_source
driver.quit()
```

```python
world_selection = BeautifulSoup(page_source, "lxml")
print(world_selection.prettify())
```

```python
noticedDiv= world_selection.find_all("div", class_= "tv-widget-idea js-userlink-popup-anchor")
noticedDiv
text=[item.find("div", class_= "tv-widget-idea__title-row").get_text("", strip = True) for item in noticedDiv]
text
text2=[item.find("p", class_= "tv-widget-idea__description-row tv-widget-idea__description-row--clamped js-widget-idea__popup").get_text("", strip = True) for item in noticedDiv]
text2
```

```python
df = pd.DataFrame(columns = ["Tittle", "Info",],
                          dtype = object)

df["Tittle"] = text
df["Info"] = text2
print(df)
```
